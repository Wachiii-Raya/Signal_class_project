```python
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
import numpy as np
import pandas as pd
import wave
import pywt
import sys
```

```python
input_data = read("71.wav")
input_data = input_data[1]
#cut 
```

```python
#len the size of signal 
size_of_signal = len(input_data)
print("Size of signal: ", size_of_signal)
```

```python
audio = input_data[1]
# plot the first 1024 samples
plt.plot(audio[0:50000])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Respiratory Sound")
# display the plot
plt.show()
```

```python
signal = np.array(audio)
fft = np.fft.fft(signal)
# Compute the absolute value to get the magnitude of each frequency component
fft_abs = np.abs(fft)

# Generate frequencies
frequencies = np.fft.fftfreq(len(fft))

# Plot
plt.figure(figsize=(7, 5))
plt.plot(frequencies, fft_abs)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Domain Representation')
plt.grid(True)
plt.show()
```
